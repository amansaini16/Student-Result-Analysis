__Distribution of Exam Scores__

The plot shows how scores are distributed for Math, Reading, and Writing. It allows you to compare:

1. __Score Distribution__: Visualize the range and frequency of scores in each subject.
2. __Score Concentration__: Identify if scores are more clustered around certain values or spread out.
3. __Subject Comparison__: See differences in score patterns between Math, Reading, and Writing.

This comparison helps to understand which subjects might be easier or harder for students and how scores vary across these subjects.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['MathScore'], bins=20, kde=True, color='blue', label='Math')
sns.histplot(df['ReadingScore'], bins=20, kde=True, color='green', label='Reading')
sns.histplot(df['WritingScore'], bins=20, kde=True, color='red', label='Writing')
plt.legend()
plt.title('Distribution of Exam Scores')
plt.show()
```

__Gender Distribution__

```python
plt.figure(figsize=(5, 5))
ax = sns.countplot(data=df, x='Gender')
ax.bar_label(ax.containers[0])  
plt.title('Count of Students by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#From above we conclude that the no.of female students is more than male students
```

__Parent Education Impact__

The code creates a heatmap that displays the average Math, Reading, and Writing scores based on the parents' education levels. It allows you to:

1. __Visualize Averages__: See the average scores for each subject grouped by different levels of parental education.
2. __Compare Impact__: Determine how the level of parental education affects student scores in each subject.
3. __Identify Patterns__: Spot trends or patterns in student performance related to parental education.

This helps to analyze the influence of parental education on student achievement in different subjects.

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean',"WritingScore": 'mean'})
```

```python
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.title("Parent Education Impact On Student Scores")
plt.show()
```

```python
#From above we have conclude that education of parent have a good impact on student scores
```

# Distribution of Ethnic Groups

```python
print(df["EthnicGroup"].unique())
```

The code creates a pie chart that shows the proportion of students in each ethnic group (A, B, C, D, E). It helps you:

1. __Visualize Proportions__: See the percentage of students in each ethnic group.
2. __Compare Groups__: Easily compare the size of each ethnic group relative to the others.
3. __Identify Dominance__: Quickly identify which ethnic group has the largest or smallest representation.

This chart provides a clear view of the distribution of students across different ethnic groups.

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels=l,autopct= "%1.2f%%")
plt.show()
```

__Comparison By Lunch Type__ 

```python
# Grouping by lunch type and comparing average scores
gb2 = df.groupby('LunchType')[['MathScore', 'ReadingScore', 'WritingScore']].mean()
print(gb2)
```

```python
sns.heatmap(gb2, annot=True, cmap="coolwarm")
plt.title("Scores Comparison by Lunch Type")
plt.show()
```
